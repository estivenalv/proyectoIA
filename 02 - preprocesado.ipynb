# Limpieza del data frame
para la limpieza del dataframe inicialmente cree una copia del data frame para poder trabajarlo sin modificar el original, empece organizando columna a columna, dependiendo de si se deben rellenar datos nulos, o algun cambio adicional.

```python
# copia del data frame
df_01 = df.copy()
```

```python
df_01.columns
```

```python
# cambiar la columna ID como un indice o un identificador unico
df_01 = df_01.set_index("ID")
df_01.head()
```

```python
# separar la columna 'periodo' en año y periodo
df_01['AÑO'] = df_01['PERIODO'].astype(str).str[:4]
df_01['PERIODO'] = df_01['PERIODO'].astype(str).str[-1]
# manteniendo estas nuevas columnas como enteros
df_01['AÑO'] = df_01['AÑO'].astype(int)
df_01['PERIODO'] = df_01['PERIODO'].astype(int)
df_01[['AÑO', 'PERIODO']].head()
```

```python
# hacer un diccionario para la columna 'ESTU_VALORMATRICULAUNIVERSIDAD' con un valor medio del rango presente
ESTU_VALORMATRICULAUNIVERSIDAD = {'Entre 1 millón y menos de 2.5 millones': 1.75,
    'Entre 2.5 millones y menos de 4 millones': 3.25,
    'Menos de 500 mil': 0.25,
    'Entre 500 mil y menos de 1 millón': 0.75,
    'Entre 4 millones y menos de 5.5 millones': 4.75,
    'Más de 7 millones': 7.75,
    'Entre 5.5 millones y menos de 7 millones': 6.25,
    'No pagó matrícula': 0,}

# con el diccionario definido se puede realizar el mapeo
df_01['ESTU_VALORMATRICULAUNIVERSIDAD'] = df_01['ESTU_VALORMATRICULAUNIVERSIDAD'].map(ESTU_VALORMATRICULAUNIVERSIDAD)

# completar los valores nulo con la media
media_mapeo1 = df_01['ESTU_VALORMATRICULAUNIVERSIDAD'].mean()
df_01['ESTU_VALORMATRICULAUNIVERSIDAD'] = df_01['ESTU_VALORMATRICULAUNIVERSIDAD'].fillna(media_mapeo1)

df_01[['ESTU_VALORMATRICULAUNIVERSIDAD']].head()
```

```python
# hacer un diccionario para la columna 'ESTU_HORASSEMANATRABAJA	' con un valor medio del rango presente
ESTU_HORASSEMANATRABAJA = {'0':0,
          'Menos de 10 horas':5,
          'Entre 11 y 20 horas':15.5,
          'Entre 21 y 30 horas':25.5,
          'Más de 30 horas':45}

# con el diccionario definido se puede realizar el mapeo
df_01['ESTU_HORASSEMANATRABAJA'] = df_01['ESTU_HORASSEMANATRABAJA'].map(ESTU_HORASSEMANATRABAJA)

# completar los valores nulos con la media
media_mapeo2 = df_01['ESTU_HORASSEMANATRABAJA'].mean()
df_01['ESTU_HORASSEMANATRABAJA'] = df_01['ESTU_HORASSEMANATRABAJA'].fillna(media_mapeo2)

df_01[['ESTU_HORASSEMANATRABAJA']].head()
```

```python
# hacer un diccionario para la columna 'FAMI_ESTRATOVIVIENDA_	' eliminando la cadena de texto y dando valores enteros
FAMI_ESTRATOVIVIENDA_ = {'Sin Estrato':0,
                         'Estrato 1':1, 'Estrato 2':2,
                         'Estrato 3':3, 'Estrato 4':4,
                         'Estrato 5':5, 'Estrato 6':6}
df_01['FAMI_ESTRATOVIVIENDA']=df_01['FAMI_ESTRATOVIVIENDA'].map(FAMI_ESTRATOVIVIENDA_)

# para poder darle un valor a los valores nulos en este caso, use condiciones, donde tuve en cuenta otra columna que pueda estar
# relacionada con el estrato de alguna forma, como lo puede ser el valor de la matricula del estudiante
estrato1 = {0, 0.25, 0.75}
estrato5 = {4.75, 6.25}
def asignar_estrato(row):
    if pd.isna(row['FAMI_ESTRATOVIVIENDA']):
        if row['ESTU_VALORMATRICULAUNIVERSIDAD'] in estrato1:
            return 1
        elif row['ESTU_VALORMATRICULAUNIVERSIDAD'] == 1.75:
            return 2
        elif row['ESTU_VALORMATRICULAUNIVERSIDAD'] == 3.25:
            return 3
        elif row['ESTU_VALORMATRICULAUNIVERSIDAD'] == df_01['ESTU_VALORMATRICULAUNIVERSIDAD'].mean():
            return 4
        elif row['ESTU_VALORMATRICULAUNIVERSIDAD'] in estrato5:
            return 5
        elif row['ESTU_VALORMATRICULAUNIVERSIDAD'] == 7.75:
            return 6
    return row['FAMI_ESTRATOVIVIENDA']

df_01['FAMI_ESTRATOVIVIENDA'] = df_01.apply(asignar_estrato, axis=1)

# si queda algun valor nulo que no responda a la función se le dara un valor de forma manual
df_01.fillna({'FAMI_ESTRATOVIVIENDA': 2}, inplace=True)
df_01['FAMI_ESTRATOVIVIENDA'].head()
```

```python
# hacer un diccionario para la columna 'FAMI_TIENEINTERNET_' al ser solo si o no es mucho mas util
FAMI_TIENEINTERNET_ = {'Si':1,'No':0}
df['FAMI_TIENEINTERNET']=df['FAMI_TIENEINTERNET'].map(FAMI_TIENEINTERNET_)

# para rellenar los valores nulos le dare el valor de forma manual
df_01.fillna({'FAMI_TIENEINTERNET': 1}, inplace=True)
df_01['FAMI_TIENEINTERNET'].head()
```

```python
# hacer un diccionario para la columna 'ESTU_PAGOMATRICULAPROPIO' al ser solo si o no es mucho mas util
ESTU_PAGOMATRICULAPROPIO = {'Si':1,'No':0}
df['ESTU_PAGOMATRICULAPROPIO']=df['ESTU_PAGOMATRICULAPROPIO'].map(ESTU_PAGOMATRICULAPROPIO)

# para rellenar los valores nulos le dare el valor de forma manual
df_01.fillna({'ESTU_PAGOMATRICULAPROPIO': 0}, inplace=True)
df_01['ESTU_PAGOMATRICULAPROPIO'].head()
```

```python
#  hacer un diccionario para la columna 'RENDIMIENTO_GLOBAL_'
RENDIMIENTO_GLOBAL_ = {'medio-bajo':0,'bajo':1,
                       'medio-alto':2,'alto':3}
df_01['RENDIMIENTO_GLOBAL']=df_01['RENDIMIENTO_GLOBAL'].map(RENDIMIENTO_GLOBAL_)
df_01['RENDIMIENTO_GLOBAL'].head()
```
